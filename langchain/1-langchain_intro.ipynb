### Streaming and Batch

#### Streaming
Most models can stream their output content while it being generated. This improves user experince, particularly for longer responses. This can be done by calling stream() method. We can use loops to process each chunk in real-time

```python
model.invoke("Write a 1000 word paragraph on AI")
```

```python
for chunk in model.stream("Write a 1000 word paragraph on AI"):
    print(chunk.text, end="|", flush="True")
```

#### Batch

Batching a collection of independent requests to a model can imporve performance and reduce costs, as the processing can be done in parallel

```python
responses = model.batch(
    [
        "Hi, How are you?",
        "What is the capital of India?",
        "What is the capital of Telangana?"
    ], config={
        'max_concurrency':5 # maximum of 5 parallel requests
    }
)

for response in responses:
    print(response.text)
```
