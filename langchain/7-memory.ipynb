```python
import os
os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")

from langchain_groq import ChatGroq
model = ChatGroq(model="openai/gpt-oss-120b")
model
```

```python
from langchain.agents import create_agent
from langgraph.checkpoint.memory import InMemorySaver

agent = create_agent(
    model= model, checkpointer=InMemorySaver()
)
```

```python
### Run with thread ID

config = {"configurable":{"thread_id":"Balaji"}}
```

```python
from langchain_core.messages import HumanMessage,SystemMessage

response = agent.invoke({"messages":[HumanMessage("Hi My name is Balaji")]},config=config)
response
```

```python
response = agent.invoke({"messages":[HumanMessage("Hi What is my name?")]},config=config)
response
```
