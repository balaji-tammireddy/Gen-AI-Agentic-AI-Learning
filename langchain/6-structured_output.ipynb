### Pydantic

```python
import os
os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")

from langchain_groq import ChatGroq

model = ChatGroq(model="openai/gpt-oss-120b")
model
```

```python
from pydantic import BaseModel, Field

class Movie(BaseModel):
    title:str=Field(description="The tite of the movie")
    year:str=Field(description="The year moview was released")
    director:str=Field(description="The director of the movie")
    rating:float=Field(description="The movie rating out of 10")
```

```python
model_with_structure = model.with_structured_output(Movie)
model_with_structure
```

```python
model.invoke("Give me the details about varanasi movie")
```

```python
model_with_structure.invoke("Give me the details about varanasi movie")
```

### Message output along with parsed structure

```python
from pydantic import BaseModel, Field

class Movie(BaseModel):
    title:str=Field(description="The tite of the movie")
    year:str=Field(description="The year moview was released")
    director:str=Field(description="The director of the movie")
    rating:float=Field(description="The movie rating out of 10")
```

```python
model_with_structure = model.with_structured_output(Movie, include_raw=True)
```

```python

response = model_with_structure.invoke("Give me details about Peddi movie")
response
```

### Nested Structure

```python
from pydantic import BaseModel,Field

class Actor(BaseModel):
    actor:str
    role:str

class MovieDetails(BaseModel):
    title:str
    year:int
    cast:list[Actor]
    genre:list[str]
    budget:float | None = Field(None, description="Budget in INR")
    collections:float | None = Field(None, description="Collections in INR")
```

```python
model_with_structure = model.with_structured_output(MovieDetails)
model_with_structure
```

```python
response = model_with_structure.invoke("Give me the details of movie RRR")
response
```
